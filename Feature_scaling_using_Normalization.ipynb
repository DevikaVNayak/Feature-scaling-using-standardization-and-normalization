```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/loan_data_set.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
sns.displot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("before")
sns.displot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("after")
sns.displot(dataset["CoapplicantIncome_min"])
plt.show()
```
