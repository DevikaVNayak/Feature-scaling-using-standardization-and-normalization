```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Feature scaling 2 types 1)Standardization
                         2)Normalizarttion

```python
dataset=pd.read_csv("/content/loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.displot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
st.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(st.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.displot(dataset["ApplicantIncome_ss"])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("before")
sns.displot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.displot(dataset["ApplicantIncome_ss"])

plt.show()
```
